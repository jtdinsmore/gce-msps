```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

plt.style.use('jcap')

DIRECTORY = "position"
if DIRECTORY == "position":
    FILE_NAME = "flux-hist-DR2.txt"
else:
    FILE_NAME = "flux-hist.txt"
CM_PER_KILOPARSEC = 3.086e21
PI = 3.1415926535
F_GCE = 1.794113925439598e-09
FGL_VERSION = 22
ERGS_PER_GEV = 0.00160218
NORMAL_AMOUNT = 47
ONE_PHOTON_FLUX = 1.9063811e-13 # erg/cm^2/s
```

```python
plt.style.use('jcap')
def get_point(xs, ys, point):
    if xs[0] > point or point > xs[-1]:
        return np.nan
    for i, x in enumerate(xs):
        if xs[i+1] > point:
            frac = (point - x) / (xs[i+1] - x)
            return frac * (ys[i+1] - ys[i]) + ys[i]

def display(data_file, label, color, style, normalize=True, alpha=1):
    f = open(data_file)
    header = f.readline()
    predicted_counts = np.asarray([float(i) for i in f.readline().split(",")[:-1]])
    predicted_fluxes = np.asarray([float(i) for i in f.readline().split(",")[:-1]])
    print(np.sum(predicted_counts))
    if normalize:
        predicted_counts *= NORMAL_AMOUNT / sum(predicted_counts)
    f.close()

    plt.plot(predicted_fluxes, predicted_counts, label=label, c=color, linestyle=style, alpha=alpha)
    width = [predicted_fluxes[i+1] - predicted_fluxes[i] for i in range(len(predicted_fluxes) - 1)]
    width.append(width[-1])
    print(width / predicted_fluxes)
    #plt.bar(predicted_fluxes, predicted_counts, width=width, color=color, align='edge', alpha=alpha)
    
    plt.xlabel("Flux received (erg / cm$^2$ / s)")
    plt.ylabel("Point source count")
    plt.xscale('log')
```

```python
def get_unscaled_count(data_file):
    f = open(data_file)
    header = f.readline()
    predicted_counts = np.asarray([float(i) for i in f.readline().split(",")[:-1]])
    f.close()
    return np.sum(predicted_counts)

print("Comparisons are from the position-dependent case without the histogram breakdown.")
print(get_unscaled_count("../luminosity-models-{}/data-1x/power-law/{}".format(DIRECTORY, FILE_NAME)), "Compare 24")
print(get_unscaled_count("../luminosity-models-{}/data-1x/wavelet-fit/{}".format(DIRECTORY, FILE_NAME)), "Compare 80")
print(get_unscaled_count("../luminosity-models-{}/data-1x/log-normal/{}".format(DIRECTORY, FILE_NAME)), "Compare 107")
print(get_unscaled_count("../luminosity-models-{}/data-1x/ploeg/{}".format(DIRECTORY, FILE_NAME)), "Compare 14")
print(get_unscaled_count("../luminosity-models-{}/data-1x/gautam/{}".format(DIRECTORY, FILE_NAME)), "Compare 8.9")
print(get_unscaled_count("../luminosity-models-{}/data-1x/nptf/{}".format(DIRECTORY, FILE_NAME)), "Compare 71")
print(get_unscaled_count("../luminosity-models-{}/data-1x/bartels18/{}".format(DIRECTORY, FILE_NAME)), "Compare 24")
```
